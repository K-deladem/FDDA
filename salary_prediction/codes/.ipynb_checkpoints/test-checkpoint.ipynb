```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donn√©es depuis le fichier CSV
data = pd.read_csv('../data/salaryMulti.csv')

# Standardisation des donn√©es
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Initialisation de l'objet PCA avec le nombre de composantes voulues
n_components = 2  # Vous pouvez ajuster ce nombre en fonction de vos besoins
pca = PCA(n_components=n_components)

# Application du PCA sur les donn√©es standardis√©es
pca.fit(data_scaled)

# Transformation des donn√©es pour obtenir les composantes principales
data_pca = pca.transform(data_scaled)

# Cr√©ation d'un DataFrame pour stocker les composantes principales
columns = ['PC' + str(i+1) for i in range(n_components)]
data_pca_df = pd.DataFrame(data=data_pca, columns=columns)

# Affichage des r√©sultats
print("Variance expliqu√©e par chaque composante principale:", pca.explained_variance_ratio_)
print("Composantes principales (PC1, PC2, ...):")
print(data_pca_df.head())

# Scatter plot des composantes principales
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['Salary'], cmap='viridis', alpha=0.8)
plt.title('PCA: Composantes principales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='PRICE')
plt.grid(True)
plt.show()
```

```python
# Import the required modules 
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, \
    mean_squared_error
import numpy as np

# Load the diabetes dataset 
from sklearn.pipeline import Pipeline
#import mes donn√©es
df= pd.read_csv('../data/salaryMulti.csv')

# Division des donn√©es en ensembles d'entra√Ænement et de test
Y = df['Salary']
X = df.drop('Salary', axis=1)
```

```python
# Create a pipeline with PCA and linear regression 
pca = PCA(n_components=3)

# Keep only the first six principal components 
reg = LinearRegression()
pipeline = Pipeline(steps=[('pca', pca),
                           ('reg', reg)])

# Fit the pipeline to the data 
pipeline.fit(X, Y)

# Predict the labels for the data 
y_pred = pipeline.predict(X) 
```

```python
# Compute the evaluation metrics 
mae = mean_absolute_error(Y, y_pred)
mse = mean_squared_error(Y, y_pred)
rmse = np.sqrt(mse)
r2 = pipeline.score(X, Y)

# Print the number of features before and after PCR 
print(f'Number of features before PCR: {X.shape[1]}')
print(f'Number of features after PCR: {pca.n_components_}')

# Print the evaluation metrics 
print(f'MAE: {mae:.2f}')
print(f'MSE: {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'R^2: {r2:.2f}') 
```
